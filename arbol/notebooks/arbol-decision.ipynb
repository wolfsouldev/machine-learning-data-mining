```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar datos
df = pd.read_csv("../data/estudiantes.csv")

# Transformar datos
df["Aprueba"] = df["Aprueba"].map({"Sí": 1, "No": 0})

# Features y etiqueta
X = df[["HorasEstudio", "Asistencia", "TareasEntregadas"]]
y = df["Aprueba"]

# Separar en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)

# Visualizar el árbol
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=X.columns, class_names=["No", "Sí"], filled=True)
plt.title("Árbol de Decisión - Predicción de Aprobación")
plt.show()
```
